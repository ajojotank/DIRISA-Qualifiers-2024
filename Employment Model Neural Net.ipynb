### Importing Libraries

### Importing data

#### Create a dictionary of data frames with the new names

```python
dataframes = {}

for year in range(2012, 2023):
    file_name = f"{year}.csv"
    file_path = f'./Data/{file_name}'
    
    data = pd.read_csv(file_path, encoding='latin-1', low_memory=False)
    dataframes[year] = data
    
    display(Markdown(f"**Column Headings for {year}:**"))
    column_names = ', '.join(data.columns)
    display(data.head())
    display(Markdown(f"`{column_names}`"))
    print("\n")
```

## Key Features to Combine and Include

### Demographic Information

- **Age**: `Age`, `age_grp`, `age_grp1` â€“ Combine these into a unified "Age" and "Age Group" feature.
- **Gender**: `Gender`, `Sex` â€“ Standardize the naming to "Gender."
- **Population Group**: `Race`, `Population` â€“ Standardize to "Population Group."
- **Geography**: `GeoType`, `Metro`, `Metro_code`, `Prov` â€“ Consolidate into "Geographic Location" or "Urban/Rural."

### Health Status

- **Chronic Illnesses**: `HLT_CHRONIC_*`, `HLT_GENHEALTH`, `HLT_MEDI` â€“ Combine into a composite feature "Chronic Health Conditions" or create separate indicators for significant illnesses.
- **Disabilities**: `DSB_*`, `disab`, `disabs`, `Sevdisab`, `SOC_DISB_PERM` â€“ Consolidate into a "Disability Status" feature.

### Education

- **Education Level**: `Education`, `Q16HIEDU`, `q16hiedu` â€“ Standardize to "Highest Education Level."
- **Current Education Status**: `EDU_ATTEND`, `Q110ATTE`, `Q112EDUI` â€“ Create a feature "Currently Attending School" with the type of institution if relevant.
- **School/Training Attendance Issues**: `EDU_ABS`, `EDU_ABS_RSN`, `EDU_ABS_DAYS` â€“ A derived feature "School Attendance Issues" might be useful.

### Economic Activity

- **Employment Status**: `employ_Status1`, `employ_Status2`, `LAB_WGE`, `LAB_VOL`, `LAB_BUS`, `LAB_LOOK` â€“ These need to be combined into a unified "Employment Status" feature.
- **Income and Salary**: `LAB_SALARY`, `LAB_SALC`, `LAB_STO`, `LAB_SALPER` â€“ Consolidate into an "Income Level" feature.
- **Sector of Employment**: `LAB_SECTOR`, `Q45SEC` â€“ Standardize to "Employment Sector."
- **Unemployment and Job Seeking**: `LAB_LOOK`, `LAB_RNSW`, `LAB_ACCJOB` â€“ Create features for "Unemployed Seeking Work" and "Not Seeking Work" based on the reasons provided.

### Transportation and Commute

- **Commute Details**: `LAB_MINS`, `LAB_TRANSP`, `Q115aMOD`, `Q44ATM` â€“ Consolidate these into "Commute Time" and "Commute Mode."

### Social and Economic Support

- **Social Grants**: `SOC_GRANT`, `SOC_GRANT_TYPE`, `SOC_GRANT_*` â€“ Combine into a "Social Grants Received" feature.
- **Household Relationships**: `HHC_RELATIONSHIP`, `HHC_MARITAL`, `HHC_FATH_*`, `HHC_MOTH_*` â€“ These could be combined into "Household Head Relationship" and "Marital Status" features.

## Cleaning 2012

```python
# Define the columns to keep initially
initial_fields = [
    'Age', 'Age_grp', 'Gender', 'Race', 'Prov', 'GeoType', 'Metro',
    'Q128aasm', 'Q128adiab', 'Q128acan', 'Q128ahiv', 'Q128ahbp', 'Q128arth', 'Q128aoth',
    'disab', 'Sevdisab', 'undisab',
    'Q16hiedu', 'Q110atte',
    'Q124aabs', 'Q124bDAY', 'Q124crsn',
    'Q21awge', 'Q21bbus', 'Q21cvol', 'Q21dret', 'Q22asto', 'Q22Msal', 'Q25sec',
    'Q26aLOOK', 'Q26cRNSW',
    'Q24atm', 'Q24bmin',
    'Q132asg', 'Q132boag', 'Q132bdis', 'Q132bcsg', 'Q132bcar', 'Q132bfos', 'Q132bwvt', 'Q132bgrn', 'Q132bsoc',
    'Q11relsh', 'Q12amarst', 'Q12bspou'
]

# Load the data from the DataFrame
df = dataframes[2012][initial_fields]

# Combine chronic condition columns into one "Chronic Condition" column
chronic_columns = ['Q128aasm', 'Q128adiab', 'Q128acan', 'Q128ahiv', 'Q128ahbp', 'Q128arth', 'Q128aoth']
df['Chronic Condition'] = df[chronic_columns].apply(lambda x: any(x == 'Yes'), axis=1)

# Function to categorize education levels
def categorize_education(level):
    if pd.isna(level):
        return 'Unknown'
    level = str(level)  # Ensure level is string to prevent errors
    if 'No schooling' in level:
        return 'No Schooling'
    elif any(f'Grade {str(digit)}' in level for digit in range(1, 6)):
        return 'Primary Education'
    elif any(f'Grade {str(digit)}' in level for digit in range(6, 8)):
        return 'Lower Secondary Education'
    elif any(f'Grade {str(digit)}' in level for digit in range(8, 10)):
        return 'Upper Secondary Education'
    elif any(f'Grade {str(digit)}' in level for digit in range(10, 12)):
        return 'Further Secondary Education'
    elif 'Grade 12' in level or 'Matric' in level:
        return 'Matric'
    elif 'Diploma' in level:
        return 'Diploma'
    elif 'Certificate' in level:
        return 'Certificate'
    elif 'Bachelors' in level:
        return 'Bachelors'
    elif 'Honours' in level or 'Higher Diploma' in level:
        return 'Honours'
    elif 'Masters' in level or 'Doctorate' in level:
        return 'Postgraduate'
    else:
        return 'Other'

df['Education Level'] = df['Q16hiedu'].apply(categorize_education)

# Combine commute details
df['Commute'] = df.apply(lambda x: f"{x['Q24atm']} - {x['Q24bmin']} mins", axis=1)

# Combine social grants into one comprehensive field
social_grant_cols = ['Q132asg', 'Q132boag', 'Q132bdis', 'Q132bcsg', 'Q132bcar', 'Q132bfos', 'Q132bwvt', 'Q132bgrn', 'Q132bsoc']
df['Social Grants'] = df[social_grant_cols].apply(lambda x: ', '.join(filter(pd.notna, x)), axis=1)

# Define function to categorize employment status
def categorize_employment(wage, bus, vol, ret):
    if wage == 'Yes':
        return 'Employed'
    elif bus == 'Yes':
        return 'Self-employed'
    elif vol == 'Yes':
        return 'Volunteering'
    elif ret == 'Yes':
        return 'Returning to work'
    else:
        return 'Unemployed'

# Apply the function to categorize employment status
df['Employment Status'] = df.apply(lambda x: categorize_employment(x['Q21awge'], x['Q21bbus'], x['Q21cvol'], x['Q21dret']), axis=1)

# Select demographic fields to keep
demographic_fields = ['Age', 'Gender', 'Race', 'Prov', 'GeoType', 'Employment Status']

# Drop fields that have been processed or are no longer needed
fields_to_drop = list(set(initial_fields) - set(demographic_fields))
df.drop(fields_to_drop, axis=1, inplace=True)

def normalize_commute(commute):
    if pd.isna(commute):
        return 'Not Applicable', 'Not Applicable'
    
    # Split the commute information into mode and time
    parts = commute.split(' - ')
    mode = parts[0].strip()
    time = parts[1].strip() if len(parts) > 1 else 'Unspecified'

    # Normalize transportation modes
    mode_mapping = {
        'Own car/other private vehicle/company vehicle': 'Private Vehicle',
        'Minibus taxi/sedan taxi/bakkie taxi': 'Taxi',
        'Bicycle/motorcycle': 'Bicycle/Motorcycle',
        'Office is at home': 'Home Office',
        'Not applicable': 'Not Applicable',
        'Other (specify)': 'Other',
        'Unspecified': 'Unspecified'
    }
    normalized_mode = mode_mapping.get(mode, 'Other')

    # Normalize commute times
    time_mapping = {
        'Less than 15 minutes': '<15 minutes',
        '15Â–30 minutes': '15-30 minutes',
        '31Â–60 minutes': '31-60 minutes',
        '61Â–90 minutes': '61-90 minutes',
        'More than 90 minutes': '>90 minutes',
        'Not applicable': 'Not Applicable',
        'Unspecified': 'Unspecified',
        'Do not know': 'Unknown'
    }
    normalized_time = time_mapping.get(time, 'Unknown')

    return normalized_mode, normalized_time

# Apply normalization function to the Commute column
df['Transportation Mode'], df['Commute Time'] = zip(*df['Commute'].apply(normalize_commute))

def normalize_social_grants(grants):
    # Split the grants string into individual responses
    responses = grants.split(', ')
    # Check if 'Yes' is in the responses to determine if any grant is received
    if 'Yes' in responses:
        return 'Yes'
    else:
        return 'No'

# Apply normalization function to the Social Grants column
df['Social Grants'] = df['Social Grants'].apply(normalize_social_grants)

# Optionally, drop the original Commute column if no longer needed
df.drop('Commute', axis=1, inplace=True)
df.drop('Commute Time', axis=1, inplace=True)
df.head()
```
