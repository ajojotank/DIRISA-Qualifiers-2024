```python
import pandas as pd
from statsmodels.tsa.holtwinters import SimpleExpSmoothing
import matplotlib.pyplot as plt

# Assuming df is your DataFrame with historical data
df = south_africa_long

# Fit the SES model
model = SimpleExpSmoothing(df['y'])
fit = model.fit(optimized=True)

# Define the number of future periods
future_periods = 5

# Generate future dates
last_date = df['ds'].max()
future_dates = pd.date_range(start=last_date, periods=future_periods + 1, freq='Y')[1:]

# Create a DataFrame for future dates
future_df = pd.DataFrame({'ds': future_dates})

# Forecast future values
forecast = fit.forecast(steps=future_periods)

# Add the forecast to the future DataFrame
future_df['forecast'] = forecast

# Combine historical data with future forecast
combined_df = pd.concat([df, future_df], ignore_index=True)

# Plot the results
plt.figure(figsize=(12, 6))
plt.plot(df['ds'], df['y'], label='Historical Data', marker='o', color='blue')
plt.plot(future_df['ds'], future_df['forecast'], label='SES Forecast', linestyle='--', color='red')

# Adding plot aesthetics
plt.title('Single Exponential Smoothing Forecast')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA

# Assuming 'south_africa_long' DataFrame has 'ds' as the date column and 'y' as the data to predict
# Example: south_africa_long = pd.DataFrame({'ds': pd.date_range(start='2000-01-01', periods=240, freq='M'), 'y': ...})

# Define and fit the ARIMA model
model = ARIMA(south_africa_long['y'], order=(5, 1, 0))  # Adjust order (p, d, q) as needed
fit = model.fit()

# Forecasting up to 2030
forecast_period = '2030-12-31'
num_forecast_periods = len(pd.date_range(start=south_africa_long['ds'].max(), end=forecast_period, freq='M'))
forecast = fit.forecast(steps=num_forecast_periods)

# Create a DataFrame for future dates and forecasted values
future_dates = pd.date_range(start=south_africa_long['ds'].max() + pd.offsets.MonthEnd(), periods=num_forecast_periods, freq='M')
forecast_df = pd.DataFrame({'ds': future_dates, 'forecast': forecast})

# Combine the original data with the forecasted data for visualization
combined_df = pd.concat([south_africa_long, forecast_df], ignore_index=True)

# Plot the original data and the ARIMA forecast
plt.figure(figsize=(12, 6))
plt.plot(south_africa_long['ds'], south_africa_long['y'], label='Original Data', marker='o')
plt.plot(future_dates, forecast, label='ARIMA Forecast to 2030', linestyle='--', color='red')
plt.title('ARIMA Forecast to 2030')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate')
plt.legend()
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.holtwinters import ExponentialSmoothing

# Assuming 'south_africa_long' DataFrame has 'ds' as the date column and 'y' as the data to predict
# Example: south_africa_long = pd.DataFrame({'ds': pd.date_range(start='2000-01-01', periods=240, freq='Y'), 'y': ...})

# Convert 'ds' to datetime format if it's not already
south_africa_long['ds'] = pd.to_datetime(south_africa_long['ds'])

# Define the Holt-Winters model
model = ExponentialSmoothing(south_africa_long['y'], trend='add', seasonal='add', seasonal_periods=12)

# Fit the model
fit = model.fit()

# Forecast future values up to 2030
forecast_period = '2030-12-31'
num_forecast_periods = len(pd.date_range(start=south_africa_long['ds'].max(), end=forecast_period, freq='Y'))
forecast = fit.forecast(steps=num_forecast_periods)

# Create a DataFrame to hold future dates and their forecasted values
future_dates = pd.date_range(start=south_africa_long['ds'].max() + pd.offsets.YearEnd(), periods=num_forecast_periods, freq='Y')
forecast_df = pd.DataFrame({'ds': future_dates, 'forecast': forecast})

# Combine the original data with the forecasted data
combined_df = pd.concat([south_africa_long[['ds', 'y']], forecast_df], ignore_index=True)

# Plot the original data and the Holt-Winters forecast
plt.figure(figsize=(12, 6))
plt.plot(south_africa_long['ds'], south_africa_long['y'], label='Original Data', marker='o')
plt.plot(future_dates, forecast, label='Holt-Winters Forecast to 2030', linestyle='--', color='red')
plt.title('Holt-Winters Forecast for Yearly Unemployment Rates up to 2030')
plt.xlabel('Year')
plt.ylabel('Unemployment Rate')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.holtwinters import ExponentialSmoothing

# Convert 'ds' to datetime format if it's not already
south_africa_long['ds'] = pd.to_datetime(south_africa_long['ds'])

# Define the Holt-Winters model without seasonal components
model = ExponentialSmoothing(south_africa_long['y'], trend='add', seasonal=None)

# Fit the model
fit = model.fit()

# Get the fitted values (in-sample predictions)
south_africa_long['fitted'] = fit.fittedvalues

# Forecast future values up to 2030
forecast_period = '2030-12-31'
num_forecast_periods = len(pd.date_range(start=south_africa_long['ds'].max() + pd.offsets.YearEnd(), end=forecast_period, freq='Y'))
forecast = fit.forecast(steps=num_forecast_periods)

# Create a DataFrame to hold future dates and their forecasted values
future_dates = pd.date_range(start=south_africa_long['ds'].max() + pd.offsets.YearEnd(), periods=num_forecast_periods, freq='Y')
forecast_df = pd.DataFrame({'ds': future_dates, 'y': forecast})

# Combine the original data with the forecasted data
combined_df = pd.concat([south_africa_long[['ds', 'y']], forecast_df], ignore_index=True)

# Plot the original data, fitted values, and the Holt-Winters forecast
plt.figure(figsize=(12, 6))
plt.plot(south_africa_long['ds'], south_africa_long['y'], label='Original Data', marker='o')
plt.plot(south_africa_long['ds'], south_africa_long['fitted'], label='Fitted Values', linestyle='-', color='blue')
plt.plot(forecast_df['ds'], forecast_df['y'], label='Holt-Winters Forecast to 2030', linestyle='--', color='red')
plt.title('Holt-Winters Forecast for Yearly Unemployment Rates (Including Fitted Values)')
plt.xlabel('Year')
plt.ylabel('Unemployment Rate')
plt.legend()
plt.show()
```
