```python
datasets = [data_2014, data_2015, data_2016, data_2017, data_2018, data_2021, data_2022]
names = ['2014', '2015', '2016', '2017', '2018', '2021','2022']
def duplicates(data, file_name):
  duplicate_count = data.duplicated().sum()
  print(f"Number of duplicate rows in {file_name}: {duplicate_count}")
  print('===============================================================================')

for dataset, name in zip(datasets, names):
  duplicates(dataset, name)
```

From the output we got above, there are no duplicates in the datasets we are working with. Now we are going to check if there are any null values in the datasets.

```python
def nulls(data, file_name):
  null_count = data.isnull().sum()
  print(f"Number of null rows in {file_name}: {null_count}")
  print('===============================================================================')

for dataset, name in zip(datasets, names):
  nulls(dataset, name)
```

From the output, we can see that there are no null values. This means that we will not have to remove duplicates and nulls during our data cleaning process. We will be moving to feature selection.

```python
def check_column_availability(datasets, target_column):
    availability_dict = {}

    for dataset, dataset_name in zip(datasets, dataset_names):
        availability_dict[dataset_name] = target_column in dataset.columns

    return availability_dict

# Example usage:
column_to_check = 'employ_Status1'  # Replace with the actual column name
availability_result = check_column_availability(all_datasets, column_to_check)

# Display the results
for dataset_name, is_available in availability_result.items():
    print(f"{dataset_name}: {column_to_check} {'is' if is_available else 'is not'} available.")
```
