## Key Features to Combine and Include

### Demographic Information

- **Age**: `Age`, `age_grp`, `age_grp1` – Combine these into a unified "Age" and "Age Group" feature.
- **Gender**: `Gender`, `Sex` – Standardize the naming to "Gender."
- **Population Group**: `Race`, `Population` – Standardize to "Population Group."
- **Geography**: `GeoType`, `Metro`, `Metro_code`, `Prov` – Consolidate into "Geographic Location" or "Urban/Rural."

### Health Status

- **Chronic Illnesses**: `HLT_CHRONIC_*`, `HLT_GENHEALTH`, `HLT_MEDI` – Combine into a composite feature "Chronic Health Conditions" or create separate indicators for significant illnesses.
- **Disabilities**: `DSB_*`, `disab`, `disabs`, `Sevdisab`, `SOC_DISB_PERM` – Consolidate into a "Disability Status" feature.

### Education

- **Education Level**: `Education`, `Q16HIEDU`, `q16hiedu` – Standardize to "Highest Education Level."
- **Current Education Status**: `EDU_ATTEND`, `Q110ATTE`, `Q112EDUI` – Create a feature "Currently Attending School" with the type of institution if relevant.
- **School/Training Attendance Issues**: `EDU_ABS`, `EDU_ABS_RSN`, `EDU_ABS_DAYS` – A derived feature "School Attendance Issues" might be useful.

### Economic Activity

- **Employment Status**: `employ_Status1`, `employ_Status2`, `LAB_WGE`, `LAB_VOL`, `LAB_BUS`, `LAB_LOOK` – These need to be combined into a unified "Employment Status" feature.
- **Income and Salary**: `LAB_SALARY`, `LAB_SALC`, `LAB_STO`, `LAB_SALPER` – Consolidate into an "Income Level" feature.
- **Sector of Employment**: `LAB_SECTOR`, `Q45SEC` – Standardize to "Employment Sector."
- **Unemployment and Job Seeking**: `LAB_LOOK`, `LAB_RNSW`, `LAB_ACCJOB` – Create features for "Unemployed Seeking Work" and "Not Seeking Work" based on the reasons provided.

### Social and Economic Support

- **Social Grants**: `SOC_GRANT`, `SOC_GRANT_TYPE`, `SOC_GRANT_*` – Combine into a "Social Grants Received" feature.
- **Household Relationships**: `HHC_RELATIONSHIP`, `HHC_MARITAL`, `HHC_FATH_*`, `HHC_MOTH_*` – These could be combined into "Household Head Relationship" and "Marital Status" features.

# Cleaning Each Year

## Cleaning 2012

```python
data_2012 = pd.read_csv('../Raw Data/2012.csv', encoding='latin-1')

# Define the column mappings and necessary transformations
corrected_column_mapping = {
    'Age': 'Age',                       # Age of the person in years
    'Gender': 'Gender',                 # Gender of the person
    'Race': 'Race',                     # Race group of the person
    'Prov': 'Province',                 # Province person lives in
    'disab': 'Chronic Illnesses',       # Chronic Illnesses
    'undisab': 'Disabilities',          # Disabilities
    'Q16hiedu': 'Education Level',      # Highest education level
    'Q132asg': 'Social Grants',         # Social grants
    'Q21awge': 'Employment Status'      # Employment status
}

# Extracting the data again with corrected column names
extracted_data_corrected = data_2012[list(corrected_column_mapping.keys())].rename(columns=corrected_column_mapping)

# Define transformation functions for each field
def chronic_illness(row):
    conditions = ['Q128aasm', 'Q128adiab', 'Q128acan', 'Q128ahiv', 'Q128ahbp', 'Q128arth', 'Q128aoth']
    return any(row[condition].strip().lower() == 'yes' for condition in conditions)

def disability_status(row):
    return row['undisab'].strip().lower() == 'disabled' or row['disab'].strip().lower() == 'disabled'

def social_grants_status(row):
    grants = ['Q132boag', 'Q132bdis', 'Q132bcsg', 'Q132bcar', 'Q132bfos', 'Q132bwvt', 'Q132bgrn', 'Q132bsoc']
    return any(row[grant].strip().lower() == 'yes' for grant in grants)

def employment_status(row):
    employment_fields = ['Q21awge', 'Q21bbus', 'Q21cvol', 'Q21dret']
    return 'Employed' if any(row[field].strip().lower() == 'yes' for field in employment_fields) else 'Unemployed'

def categorize_education(level):
    level = level.strip().lower()
    if "no schooling" in level or "unspecified" in level or "do not know" in level or "other" in level:
        return 'None'
    elif any(x in level for x in ["grade r/0", "grade 1", "grade 2", "grade 3", "grade 4", "grade 5", "grade 6", "grade 7"]):
        return 'Primary'
    elif any(x in level for x in ["grade 8", "grade 9", "grade 10", "grade 11", "grade 12", "matric"]):
        return 'Secondary'
    elif "degree" in level or "diploma" in level or "ntc" in level or "higher" in level or "certificate" in level:
        return 'Tertiary'
    else:
        return 'None'

# Apply transformations using the defined functions
extracted_data_corrected['Chronic Illnesses'] = data_2012.apply(chronic_illness, axis=1)
extracted_data_corrected['Disabilities'] = data_2012.apply(disability_status, axis=1)
extracted_data_corrected['Social Grants'] = data_2012.apply(social_grants_status, axis=1)
extracted_data_corrected['Employment Status'] = data_2012.apply(employment_status, axis=1)
extracted_data_corrected['Education Level'] = data_2012['Q16hiedu'].apply(categorize_education)

dataframes[2012] = extracted_data_corrected
dataframes[2012].sample(10)
```

## Cleaning 2013

```python
data_2013 = pd.read_csv('../Raw Data/2013.csv', encoding='latin-1')

# Define the column mappings and necessary transformations
column_mapping = {
    'Age': 'Age',                       # Age of the person in years
    'Gender': 'Gender',                 # Gender of the person
    'Race': 'Race',                     # Race group of the person
    'Prov': 'Province',                 # Province person lives in
    'undisab': 'Chronic Illnesses',     # Placeholder for Chronic Illnesses; need specific columns
    'disab': 'Disabilities',            # Disabilities status
    'Q16HIEDU': 'Education Level',      # Highest education level
    'Q31asg': 'Social Grants',          # Placeholder for Social Grants; need specific columns
    'Q41aWGE': 'Employment Status'      # Placeholder for Employment status; need to include other job-related columns
}

extracted_data_corrected = data_2013[list(column_mapping.keys())].rename(columns=column_mapping)

# Applying transformations
def chronic_illness(row):
    # Check any relevant chronic illness conditions
    conditions = ['Q26aASM', 'Q26aDBT', 'Q26aCAN', 'Q26aHIV', 'Q26aHBP', 'Q26aRTH', 'Q26aSTRK', 'Q26aHEART', 'Q26aTB', 'Q26aOTH']
    return any(row.get(condition, "").strip().lower() == 'yes' for condition in conditions if condition in row)

def disability_status(row):
    # Check if either undisab or disab columns indicate a disability
    return row['undisab'].strip().lower() == 'disabled' or row['disab'].strip().lower() == 'disabled'

def social_grants_status(row):
    # Check for any active social grants
    grants = ['Q31bOAG', 'Q31bDIS', 'Q31bCSG', 'Q31bCAR', 'Q31bFOS', 'Q31bWVT', 'Q31bGRN', 'Q31bSOC']
    return any(row.get(grant, "").strip().lower() == 'yes' for grant in grants if grant in row)

def employment_status(row):
    # Check for any form of employment or business activity
    employment_fields = ['Q41aWGE', 'Q41bBUS', 'Q41cVOL', 'Q41dRET']
    return 'Employed' if any(row.get(field, "").strip().lower() == 'yes' for field in employment_fields if field in row) else 'Unemployed'

def categorize_education(level):
    # Categorize education level based on the highest level achieved
    if not isinstance(level, str):
        return 'None'
    level = level.strip().lower()
    if "no schooling" in level or "unspecified" in level or "do not know" in level or "other" in level:
        return 'None'
    elif any(x in level for x in ["grade r/0", "grade 1", "grade 2", "grade 3", "grade 4", "grade 5", "grade 6", "grade 7"]):
        return 'Primary'
    elif any(x in level for x in ["grade 8", "grade 9", "grade 10", "grade 11", "grade 12", "matric"]):
        return 'Secondary'
    elif "degree" in level or "diploma" in level or "ntc" in level or "higher" in level or "certificate" in level:
        return 'Tertiary'
    else:
        return 'None'

# Apply transformations using the defined functions
extracted_data_corrected['Chronic Illnesses'] = data_2013.apply(chronic_illness, axis=1)
extracted_data_corrected['Disabilities'] = data_2013.apply(disability_status, axis=1)
extracted_data_corrected['Social Grants'] = data_2013.apply(social_grants_status, axis=1)
extracted_data_corrected['Employment Status'] = data_2013.apply(employment_status, axis=1)
extracted_data_corrected['Education Level'] = data_2013['Q16HIEDU'].apply(categorize_education)

dataframes[2013] = extracted_data_corrected
dataframes[2013].sample(10)
```

## Cleaning 2019

```python
df = pd.read_csv('../Raw Data/2019.csv', encoding='latin-1', low_memory=False)

# Mapping the relevant columns to simplified categories or converting them as needed
gender_map = {'Male': 'Male', 'Female': 'Female'}
race_map = {'African/Black': 'African/Black', 'Coloured': 'Coloured', 'Indian/Asian': 'Indian/Asian', 'White': 'White'}
province_map = {'Western Cape': 'Western Cape', 'Eastern Cape': 'Eastern Cape', 'Northern Cape': 'Northern Cape',
                'Free State': 'Free State', 'KwaZulu-Natal': 'KwaZulu-Natal', 'North West': 'North West',
                'Gauteng': 'Gauteng', 'Mpumalanga': 'Mpumalanga', 'Limpopo': 'Limpopo'}

# Define function to determine if a chronic illness is present
def chronic_illness(row):
    illnesses = ['HLT_CHRONIC_ASM', 'HLT_CHRONIC_DBT', 'HLT_CHRONIC_CAN', 'HLT_CHRONIC_HIV',
                 'HLT_CHRONIC_HBP', 'HLT_CHRONIC_ARTH', 'HLT_CHRONIC_STRK', 'HLT_CHRONIC_HEART',
                 'HLT_CHRONIC_TB', 'HLT_CHRONIC_MENTAL', 'HLT_CHRONIC_EPIL', 'HLT_CHRONIC_MEN',
                 'HLT_CHRONIC_PNEU', 'HLT_CHRONIC_BRON', 'HLT_CHRONIC_OSTER', 'HLT_CHRONIC_MAL']
    return any(row[illness] == 'Yes' for illness in illnesses)

# Define education and employment mapping functions
def map_education(level):
    if level in ['No schooling', 'Unspecified', 'DO NOT KNOW']:
        return 'None'
    elif level in ['Grade R/0', 'Grade 1/Sub A/Class 1', 'Grade 2/Sub B/Class 2', 'Grade 3/Standard 1/ABET/AET 1',
                   'Grade 4/Standard 2', 'Grade 5/Standard 3/ABET/AET 2', 'Grade 6/Standard 4', 'Grade 7/Standard 5/ABET/AET 3']:
        return 'Primary'
    elif 'Grade' in level and ('Grade 8' in level or 'Grade 9' in level or 'Grade 10' in level or 'Grade 11' in level or 'Grade 12' in level):
        return 'Secondary'
    elif 'Diploma' in level or 'Degree' in level or 'Doctoral' in level or 'Honours' in level or 'Post Higher Diploma' in level or 'NTC' in level or 'N' in level:
        return 'Tertiary'
    else:
        return 'Other'

def map_employment(status):
    if status == 'Yes':
        return 'Employed'
    elif status == 'No':
        return 'Unemployed'
    else:
        return 'NaN'

# Apply mappings and create new columns
df['Gender'] = df['Sex'].map(gender_map)
df['Race'] = df['Population'].map(race_map)
df['Province'] = df['prov'].map(province_map)
df['Chronic Illnesses'] = df.apply(chronic_illness, axis=1)
df['Disabilities'] = df['disab'].map({'Not disabled': False, 'Disabled': True})
df['Education Level'] = df['Education'].apply(map_education)
df['Social Grants'] = df['SOC_GRANT'].map({'Yes': True, 'No': False})
df['Employment Status'] = df['LAB_WGE'].apply(map_employment)

final_df = df[['age', 'Gender', 'Race', 'Province', 'Chronic Illnesses', 'Disabilities',
               'Education Level', 'Social Grants', 'Employment Status']].rename(columns={'age': 'Age'})

dataframes[2019] = final_df
dataframes[2019].sample(10)
```
